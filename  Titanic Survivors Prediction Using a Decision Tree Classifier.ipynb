## Model Configuration and Training

```python
# creating the decision tree classifier, here we use two criterions gini-index and entropy to create the decision tree classifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=1)
classifier2=DecisionTreeClassifier(criterion='gini', random_state=1)
```

```python
# fitting the decision tree classifier.
x_train=X[['Sex','Age','Fare','Embarked']]
y_train=X[['Survived']].dropna()
y_train=y_train.dropna()

x_train['Age'].fillna(x_train['Age'].mean(),inplace=True)
x_train['Fare'].fillna(x_train['Fare'].mean(),inplace=True)

```

```python
x_train,x_test,y_train,y_test=train_test_split(X[['Sex','Age','Fare','Embarked']],X['Survived'],test_size=0.3,random_state=1)
```

```python
#dropping  null values in the x_train column of the dataset
x_train['Age']=x_train['Age'].replace(np.NaN,x_train['Age'].mean())
x_test['Age']=x_test['Age'].replace(np.NaN,x_test['Age'].mean())
#dropping null values from the data
x_train['Embarked']=x_train['Embarked'].fillna(1)
x_test['Embarked']=x_test['Embarked'].fillna(1)
```

```python
# fitting both our models
by_entropy=classifier.fit(x_train,y_train)
by_gini=classifier2.fit(x_train,y_train)
```

```python
# predicting using the trained Decision Tree classifer models.
prediction_from_entropy=by_entropy.predict(x_test)
prediction_from_gini=by_gini.predict(x_test)
```

```python
# Generating Classification Reports
mat1=confusion_matrix(prediction_from_entropy,y_test.values)
mat2=confusion_matrix(prediction_from_gini,y_test.values)

report_for_gini=classification_report(prediction_from_gini,y_test.values)
report_for_entropy=classification_report(prediction_from_entropy,y_test.values)
```

```python

def findK(data,Max):
    distortion=[]
    inertia=[]
    distortionmap={}
    inertiamap={}
    for i in range(1,Max):
        kmeans=KMeans(n_clusters=i)
        kmeansModel=kmeans.fit(data)
        
        distortion.append(sum(np.min(cdist(data,kmeansModel.cluster_centers_,'euclidean'),axis=1))/len(data))
        inertia.append(kmeansModel.inertia_)
        
        
        distortionmap[i]=sum(np.min(cdist(data,kmeansModel.cluster_centers_,'euclidean'),axis=1))/len(data)
        inertiamap[i]=kmeansModel.inertia_
    
    K=range(1,Max) 
    plt.plot(K,distortion,'bx-')
    plt.legend('Distortion')
    plt.xlabel("Scores")
    plt.ylabel("Number of Clusters")
    plt.title("A plot of distortion")
    plt.show()
    
    
    plt.plot(K,inertia,'bx-',color='black')
    plt.title("A plot of inertia")
    plt.legend("Inertia")
    plt.show()
    
    
```
